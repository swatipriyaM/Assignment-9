# Apriori algorithim

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets = apriori(num_dataset, min_support=0.15, use_colnames=True,verbose=1)
print(frequent_itemsets.head())
```

```python
frequent_itemsets.sort_values("support", ascending = False).shape
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
rules= association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
print(rules.head())
```

```python
rules[(rules.support > 0.015) & (rules.confidence > 0.4)].sort_values("confidence", ascending = False).shape
```

```python
rules.sort_values('lift',ascending=False)
```

```python
rules[rules.lift>1]
```

```python

rules['lhs items'] = rules['antecedents'].apply(lambda x:len(x) )
rules[rules['lhs items']>1].sort_values('lift', ascending=False).head()
```

```python
import seaborn as sns
```

```python
rules['antecedents_'] = rules['antecedents']
# Generate a heatmap with annotations dents'].apply(lambda a: ','.join(list(a)))
rules['consequents_'] = rules['consequents'].apply(lambda a: ','.join(list(a)))
# Transform the DataFrame of rules into a matrix using the confidence metric
pivot = rules[rules['lhs items']>1].pivot(index = 'antecedents_', 
                    columns = 'consequents_', values= 'confidence')
sns.heatmap(pivot, annot = True)
plt.title('Heat Map - For Confidence Metric')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

# fp growth algorithim

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
frequent_itemsets_fp=fpgrowth(num_dataset, min_support=0.1, use_colnames=True,verbose=1)
print(frequent_itemsets_fp.shape)
```

```python

frequent_itemsets_fp.sort_values("support", ascending = False).head()
```

```python
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.8)
print(rules_fp.head())
```

```python
rules_fp[(rules_fp.support > 0.1) & (rules_fp.confidence > 0.4)].sort_values("confidence", ascending = False).shape
```

```python
rules_fp['lhs items'] = rules_fp['antecedents'].apply(lambda x:len(x) )
rules_fp[rules_fp['lhs items']>1].sort_values('lift', ascending=False).head()
```

```python
rules_fp['antecedents_'] = rules_fp['antecedents'].apply(lambda a: ','.join(list(a)))
rules_fp['consequents_'] = rules_fp['consequents'].apply(lambda a: ','.join(list(a)))
# Transform the DataFrame of rules into a matrix using the confidence metric
pivot = rules_fp[rules_fp['lhs items']>1].pivot(index = 'antecedents_', 
                    columns = 'consequents_', values= 'confidence')
```

```python
sns.heatmap(pivot, annot = True)
plt.title('Heat Map - For Confidence Metric')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```
